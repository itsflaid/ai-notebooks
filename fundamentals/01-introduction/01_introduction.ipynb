## 1. Apa itu Jupyter Notebook?

Jupyter Notebook adalah aplikasi web interaktif yang memungkinkan lo menulis dan menjalankan kode Python **langsung di browser**, bercampur dengan teks, gambar, dan visualisasi.

Kenapa populer di dunia AI/ML?
- ✅ Bisa run kode per bagian (per cell) — tidak harus run semua sekaligus
- ✅ Output langsung keliatan di bawah cell
- ✅ Bisa campurin kode, penjelasan, dan grafik dalam satu file
- ✅ File `.ipynb` bisa dibuka di GitHub, Colab, VS Code

> **Analogi:** Jupyter itu seperti buku catatan lab — lo nulis eksperimen, langsung lihat hasilnya, dan bisa tambah catatan di sampingnya.

## 3. Mode: Edit vs Command

Jupyter punya 2 mode:

**Edit Mode** (border cell biru/hijau)
- Lo lagi ngetik di dalam cell
- Masuk dengan: klik cell atau tekan `Enter`

**Command Mode** (border cell abu-abu)
- Lo bisa navigasi antar cell, ganti tipe, hapus, dll
- Masuk dengan: tekan `Esc`

> **Penting:** Banyak shortcut hanya jalan di Command Mode!

## 4. Keyboard Shortcuts Wajib Hafal

### Shortcut Universal (Edit & Command Mode)
| Shortcut | Aksi |
|----------|------|
| `Shift + Enter` | Run cell, pindah ke cell berikutnya |
| `Ctrl + Enter` | Run cell, tetap di cell yang sama |
| `Alt + Enter` | Run cell, buat cell baru di bawah |

### Shortcut Command Mode (`Esc` dulu)
| Shortcut | Aksi |
|----------|------|
| `A` | Insert cell di **atas** |
| `B` | Insert cell di **bawah** |
| `D, D` | Hapus cell (tekan D dua kali) |
| `M` | Ganti ke Markdown |
| `Y` | Ganti ke Code |
| `Z` | Undo hapus cell |
| `Shift + M` | Merge cell yang diseleksi |
| `0, 0` | Restart kernel |

### Shortcut Edit Mode
| Shortcut | Aksi |
|----------|------|
| `Ctrl + Z` | Undo |
| `Ctrl + Shift + -` | Split cell di posisi cursor |
| `Tab` | Autocomplete / indent |

## 5. Kernel — Otak di Balik Notebook

**Kernel** adalah proses Python yang berjalan di background dan mengeksekusi kode lo.

Yang perlu lo tahu:
- Semua variabel hidup di kernel selama sesi aktif
- Kalau kernel mati/restart, **semua variabel hilang** — harus run ulang dari awal
- Kernel bisa hang kalau kode lo infinite loop atau terlalu berat

**Cara kontrol kernel:**
- `Kernel → Restart` — restart kernel (variabel hilang)
- `Kernel → Restart & Run All` — restart lalu run semua cell dari atas
- `Kernel → Interrupt` — stop eksekusi yang lagi jalan (Ctrl+C)

> **Gotcha paling sering:** Lo run cell B dulu sebelum cell A, lalu restart kernel dan bingung kenapa error. Selalu pastikan urutan run cell masuk akal dari atas ke bawah!

```python
# Cek status kernel — lihat angka di sebelah kiri cell
# [1] = cell pertama yang dirun
# [*] = cell sedang running
# [ ] = belum pernah dirun

import sys
print(f"Python version: {sys.version}")
print("Kernel aktif dan siap dipakai!")
```

## 6. Markdown Dasar

Markdown cell bisa render teks formatted. Berikut syntax dasarnya:

**Heading:** `# H1`, `## H2`, `### H3`

**Bold:** `**teks**` → **teks**

**Italic:** `*teks*` → *teks*

**Code inline:** `` `kode` `` → `kode`

**List:**
- Item 1
- Item 2
  - Sub-item

**Numbered list:**
1. Pertama
2. Kedua

**Link:** `[teks](url)` → [Google](https://google.com)

**Math (LaTeX):** `$E = mc^2$` → $E = mc^2$

**Block math:** 
$$\sum_{i=0}^{n} x_i = \frac{n(n+1)}{2}$$
