## 1. Variabel dan Tipe Data

```python
# Python tidak perlu deklarasi tipe data
nama = "Jupyter"          # str
umur = 25                  # int
tinggi = 170.5             # float
aktif = True               # bool
kosong = None              # NoneType

# Cek tipe data
print(type(nama))          # <class 'str'>
print(type(umur))          # <class 'int'>
print(type(aktif))         # <class 'bool'>
```

```python
# Konversi tipe data
angka_str = "42"
angka_int = int(angka_str)    # str → int
angka_float = float(angka_str) # str → float

print(angka_int + 8)          # 50
print(str(angka_int))         # '42' (int → str)
print(bool(0), bool(1))       # False, True
```

## 2. String

```python
# String operations
teks = "Hello, Python!"

print(teks.upper())           # HELLO, PYTHON!
print(teks.lower())           # hello, python!
print(teks.replace("Python", "Jupyter"))  # Hello, Jupyter!
print(teks.split(", "))       # ['Hello', 'Python!']
print(len(teks))              # 14
print(teks[0])                # H (indexing)
print(teks[7:13])             # Python (slicing)
```

```python
# f-string (cara modern format string)
nama = "Budi"
nilai = 95.5

print(f"Nama: {nama}")
print(f"Nilai: {nilai:.1f}")   # 1 desimal
print(f"2 + 2 = {2 + 2}")     # ekspresi langsung

# Multi-line string
paragraf = """
Baris pertama
Baris kedua
Baris ketiga
"""
print(paragraf)
```

## 3. List, Tuple, Dict, Set

```python
# LIST — ordered, mutable (bisa diubah)
buah = ["apel", "mangga", "pisang"]

buah.append("jeruk")         # tambah di akhir
buah.insert(1, "anggur")     # tambah di index 1
buah.remove("mangga")        # hapus by value
item = buah.pop()            # hapus & ambil item terakhir

print(buah)
print(buah[0])               # apel
print(buah[-1])              # item terakhir
print(buah[1:3])             # slicing
```

```python
# TUPLE — ordered, immutable (tidak bisa diubah)
koordinat = (10.5, -7.8)
r, g, b = 255, 128, 0        # unpacking

print(koordinat[0])          # 10.5
print(r, g, b)

# DICT — key-value pairs
user = {
    "nama": "Budi",
    "umur": 25,
    "aktif": True
}

print(user["nama"])          # Budi
print(user.get("email", "tidak ada"))  # default value
user["email"] = "budi@mail.com"  # tambah key baru
print(user.keys())
print(user.values())
```

```python
# SET — unordered, unique values
angka = {1, 2, 3, 2, 1}     # duplikat otomatis dihapus
print(angka)                 # {1, 2, 3}

a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
print(a & b)                 # irisan: {3, 4}
print(a | b)                 # gabungan: {1, 2, 3, 4, 5, 6}
print(a - b)                 # selisih: {1, 2}
```

## 4. Kondisi (if/elif/else)

```python
nilai = 78

if nilai >= 90:
    grade = "A"
elif nilai >= 80:
    grade = "B"
elif nilai >= 70:
    grade = "C"
else:
    grade = "D"

print(f"Nilai {nilai} → Grade {grade}")

# Ternary (one-liner)
status = "lulus" if nilai >= 60 else "tidak lulus"
print(status)
```

## 5. Loop

```python
# FOR loop
buah = ["apel", "mangga", "pisang"]
for item in buah:
    print(item)

print("---")

# range()
for i in range(5):           # 0, 1, 2, 3, 4
    print(i, end=" ")
print()

# enumerate() — dapat index sekaligus
for idx, item in enumerate(buah):
    print(f"{idx}: {item}")
```

```python
# WHILE loop
counter = 0
while counter < 5:
    print(counter, end=" ")
    counter += 1
print()

# break dan continue
for i in range(10):
    if i == 3:
        continue        # skip i=3
    if i == 7:
        break           # stop di i=7
    print(i, end=" ")
print()
```

## 6. Function

```python
# Function dasar
def sapa(nama, salam="Halo"):
    """Function dengan default parameter."""
    return f"{salam}, {nama}!"

print(sapa("Budi"))            # Halo, Budi!
print(sapa("Ani", "Selamat pagi"))  # Selamat pagi, Ani!

# Multiple return values
def hitung(a, b):
    return a + b, a - b, a * b

tambah, kurang, kali = hitung(10, 3)
print(tambah, kurang, kali)    # 13, 7, 30
```

```python
# *args dan **kwargs
def jumlahkan(*angka):
    return sum(angka)

print(jumlahkan(1, 2, 3))      # 6
print(jumlahkan(1, 2, 3, 4, 5))  # 15

def tampilkan(**info):
    for key, val in info.items():
        print(f"  {key}: {val}")

tampilkan(nama="Budi", umur=25, kota="Jakarta")

# Lambda function
kuadrat = lambda x: x ** 2
print(kuadrat(5))              # 25

# Lambda dengan map()
angka = [1, 2, 3, 4, 5]
kuadrat_list = list(map(lambda x: x**2, angka))
print(kuadrat_list)            # [1, 4, 9, 16, 25]
```

## 7. List Comprehension

```python
# Cara biasa
kuadrat = []
for x in range(10):
    kuadrat.append(x**2)

# List comprehension (lebih Pythonic)
kuadrat = [x**2 for x in range(10)]
print(kuadrat)

# Dengan kondisi
genap = [x for x in range(20) if x % 2 == 0]
print(genap)

# Dict comprehension
kuadrat_dict = {x: x**2 for x in range(5)}
print(kuadrat_dict)
```

## 8. Error Handling

```python
# try/except
def bagi(a, b):
    try:
        hasil = a / b
        return hasil
    except ZeroDivisionError:
        print("Error: tidak bisa dibagi nol!")
        return None
    except TypeError as e:
        print(f"Error tipe data: {e}")
        return None
    finally:
        print("(finally selalu dijalankan)")   

print(bagi(10, 2))
print(bagi(10, 0))
print(bagi(10, "a"))
```

```python
# LATIHAN AKHIR
# Buat function yang menerima list nilai dan return statistik dasarnya

def statistik(nilai_list):
    """Return dict berisi min, max, rata-rata dari list nilai."""
    if not nilai_list:
        return None
    return {
        "min": min(nilai_list),
        "max": max(nilai_list),
        "rata_rata": sum(nilai_list) / len(nilai_list),
        "jumlah_data": len(nilai_list)
    }

nilai = [85, 92, 78, 95, 88, 73, 90]
hasil = statistik(nilai)

for key, val in hasil.items():
    print(f"{key:>15}: {val}")
```
