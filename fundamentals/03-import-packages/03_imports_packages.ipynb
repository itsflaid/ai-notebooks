## 1. Import Dasar

```python
# Import seluruh module
import math
import os
import sys

print(math.pi)           # 3.14159...
print(math.sqrt(16))     # 4.0
print(math.ceil(4.2))    # 5
print(math.floor(4.8))   # 4
```

```python
# Import dengan alias — standar di dunia data science
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sekarang bisa pakai np, pd, plt sebagai shortcut
arr = np.array([1, 2, 3, 4, 5])
print(arr)
print(arr.mean())
```

```python
# Import spesifik dari module
from math import pi, sqrt, factorial
from datetime import datetime, timedelta
from collections import Counter, defaultdict

# Langsung pakai tanpa prefix
print(pi)                      # 3.14159...
print(sqrt(25))                # 5.0
print(factorial(5))            # 120

sekarang = datetime.now()
print(sekarang.strftime("%Y-%m-%d %H:%M"))

besok = sekarang + timedelta(days=1)
print(besok.strftime("%Y-%m-%d"))
```

```python
# Import dengan alias custom
from collections import OrderedDict as ODict

data = ODict()
data['pertama'] = 1
data['kedua'] = 2
print(data)
```

## 2. Standard Library yang Sering Dipakai

```python
import os

# Navigasi filesystem
print(os.getcwd())              # direktori saat ini
print(os.listdir('.'))          # isi direktori

# Path manipulation
path = os.path.join('folder', 'subfolder', 'file.txt')
print(path)                     # folder/subfolder/file.txt

print(os.path.exists('.'))      # True (cek apakah path ada)
print(os.path.isfile('.'))      # False
print(os.path.isdir('.'))       # True
```

```python
from pathlib import Path

# pathlib — cara modern handle path (lebih readable dari os.path)
p = Path('.')
print(p.resolve())              # absolute path

# Buat path baru
file_path = Path('data') / 'raw' / 'dataset.csv'
print(file_path)                # data/raw/dataset.csv
print(file_path.suffix)        # .csv
print(file_path.stem)          # dataset
print(file_path.parent)        # data/raw

# List semua file .ipynb di direktori saat ini
notebooks = list(Path('.').glob('*.ipynb'))
print(f"Ditemukan {len(notebooks)} notebook")
```

```python
import json
import random
import time

# JSON
data = {"nama": "Budi", "nilai": [85, 92, 78]}
json_str = json.dumps(data, indent=2)   # dict → JSON string
print(json_str)

parsed = json.loads(json_str)            # JSON string → dict
print(type(parsed))

# Random
print(random.randint(1, 100))           # int random
print(random.choice(['a', 'b', 'c']))   # pilih random dari list

sample = random.sample(range(100), 5)   # 5 angka unik
print(sample)
```

```python
# Install package yang belum ada, baru import
# Pattern umum di Colab / notebook baru:

try:
    import requests
    print(f"requests sudah ada: v{requests.__version__}")
except ImportError:
    print("requests tidak ada, menginstall...")
    import subprocess
    subprocess.run(['pip', 'install', 'requests'], check=True)
    import requests
    print(f"requests berhasil diinstall: v{requests.__version__}")
```

## 4. Cek Versi Library

```python
# LATIHAN
# Import semua library yang dibutuhkan proyek data science
# dan tampilkan versinya dalam format tabel rapi

libraries = [
    ('numpy', 'np'),
    ('pandas', 'pd'),
    ('matplotlib', None),
]

print(f"{'Library':<15} {'Version':<12} {'Status'}")
print("-" * 40)

for lib_name, alias in libraries:
    try:
        lib = __import__(lib_name)
        version = getattr(lib, '__version__', 'unknown')
        print(f"{lib_name:<15} {version:<12} ✅ OK")
    except ImportError:
        print(f"{lib_name:<15} {'N/A':<12} ❌ Not installed")
```
