# 04 — Magic Commands

Magic commands adalah fitur spesial Jupyter yang tidak ada di Python biasa. Dimulai dengan `%` (line magic) atau `%%` (cell magic).

**Yang akan lo pelajari:**
- Line magic (`%`) vs Cell magic (`%%`)
- Timing: `%time`, `%timeit`
- Inspeksi: `%who`, `%whos`, `%pinfo`
- Shell: `%cd`, `%ls`, `%pwd`
- Matplotlib inline
- `%run`, `%load`, `%save`
- `%%bash`, `%%html`, `%%latex`

## 1. Timing — %time dan %timeit

```python
# %time — jalankan SEKALI, ukur waktunya
%time sum(range(1_000_000))
```

```python
# %timeit — jalankan BERKALI-KALI, ambil rata-rata (lebih akurat)
# Default: 7 run, masing-masing loop beberapa kali
%timeit sum(range(1_000_000))
```

```python
import numpy as np

# Bandingkan performa: Python list vs NumPy
python_list = list(range(1_000_000))
numpy_array = np.arange(1_000_000)

print("Python list sum:")
%timeit sum(python_list)

print("\nNumPy array sum:")
%timeit np.sum(numpy_array)
```

```python
# %%timeit — cell magic, timing seluruh cell
%%timeit
result = []
for i in range(10000):
    result.append(i ** 2)
```

```python
%%timeit
# Bandingkan dengan list comprehension
result = [i ** 2 for i in range(10000)]
```

## 5. Matplotlib Inline

```python
# Wajib dijalankan sekali sebelum plotting
# Supaya plot muncul langsung di notebook
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)
plt.figure(figsize=(8, 4))
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.legend()
plt.title('Sin dan Cos')
plt.grid(True)
plt.show()
```

```python
# %matplotlib notebook — plot interaktif (zoom, pan)
# %matplotlib widget   — versi modern (butuh ipympl)

# Set resolusi lebih tinggi
%config InlineBackend.figure_format = 'retina'

plt.figure(figsize=(6, 4))
plt.scatter(np.random.randn(100), np.random.randn(100), alpha=0.7)
plt.title('Scatter Plot')
plt.show()
```

## 6. Run dan Load File

```python
# Buat file Python sederhana dulu
script_content = '''
print("Script dijalankan!")

def hitung_luas(panjang, lebar):
    return panjang * lebar

luas = hitung_luas(5, 3)
print(f"Luas: {luas}")
'''

with open('temp_script.py', 'w') as f:
    f.write(script_content)

print("File temp_script.py berhasil dibuat")
```

```python
# %run — jalankan file Python eksternal
# Variabel dan function dari script akan tersedia di namespace
%run temp_script.py
```

```python
# Setelah %run, function dari script bisa diakses
print(hitung_luas(10, 7))    # 70
```

```python
# %save — simpan cell ke file .py
# %save nama_file 1-5     # simpan cell 1 sampai 5
# %save -a nama_file 10   # append cell 10 ke file yang ada

# %load — load kode dari file ke cell
# %load temp_script.py    # isi cell jadi kode dari file
```
