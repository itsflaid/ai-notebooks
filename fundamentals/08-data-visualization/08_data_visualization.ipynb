```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Set style default yang lebih cantik
plt.style.use('seaborn-v0_8-whitegrid')
print("Matplotlib siap!")
```

## 1. Line Chart

```python
# Line chart sederhana
x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(x, np.sin(x), label='sin(x)', color='#2196F3', linewidth=2)
ax.plot(x, np.cos(x), label='cos(x)', color='#F44336', linewidth=2, linestyle='--')
ax.plot(x, np.sin(x) * np.cos(x), label='sin·cos', color='#4CAF50', linewidth=1.5, alpha=0.8)

ax.set_title('Fungsi Trigonometri', fontsize=14, fontweight='bold')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.set_xlim(0, 10)

plt.tight_layout()
plt.show()
```

```python
# Line chart — data time series
bulan = ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun',
         'Jul', 'Agu', 'Sep', 'Okt', 'Nov', 'Des']
penjualan_a = [120, 135, 148, 162, 155, 170, 182, 175, 190, 205, 220, 250]
penjualan_b = [80, 95, 88, 102, 115, 108, 125, 130, 118, 140, 155, 165]

fig, ax = plt.subplots(figsize=(12, 5))

ax.plot(bulan, penjualan_a, 'o-', color='#2196F3', label='Produk A', linewidth=2, markersize=6)
ax.plot(bulan, penjualan_b, 's-', color='#FF9800', label='Produk B', linewidth=2, markersize=6)

ax.fill_between(range(len(bulan)), penjualan_a, alpha=0.1, color='#2196F3')
ax.set_xticks(range(len(bulan)))
ax.set_xticklabels(bulan)
ax.set_title('Penjualan Bulanan 2024', fontsize=14)
ax.set_ylabel('Unit Terjual')
ax.legend()

plt.tight_layout()
plt.show()
```

## 2. Bar Chart

```python
# Bar chart horizontal — bagus untuk perbandingan label panjang
jurusan = ['Informatika', 'Manajemen', 'Akuntansi', 'Hukum', 'Kedokteran']
mahasiswa = [450, 320, 280, 210, 180]
colors = ['#2196F3', '#4CAF50', '#FF9800', '#9C27B0', '#F44336']

fig, ax = plt.subplots(figsize=(9, 5))

bars = ax.barh(jurusan, mahasiswa, color=colors, edgecolor='white', height=0.6)

# Tambah label di ujung bar
for bar, val in zip(bars, mahasiswa):
    ax.text(val + 5, bar.get_y() + bar.get_height()/2,
            f'{val}', va='center', fontweight='bold')

ax.set_title('Jumlah Mahasiswa per Jurusan', fontsize=14)
ax.set_xlabel('Jumlah Mahasiswa')
ax.set_xlim(0, max(mahasiswa) * 1.15)

plt.tight_layout()
plt.show()
```

## 3. Histogram

```python
# Generate data
np.random.seed(42)
nilai = np.random.normal(75, 12, 500)
nilai = np.clip(nilai, 0, 100)

fig, ax = plt.subplots(figsize=(9, 5))

n, bins, patches = ax.hist(nilai, bins=30, color='#2196F3', edgecolor='white', alpha=0.8)

# Warnai berdasarkan threshold
for patch, left_edge in zip(patches, bins[:-1]):
    if left_edge < 60:
        patch.set_facecolor('#F44336')
    elif left_edge >= 80:
        patch.set_facecolor('#4CAF50')

ax.axvline(nilai.mean(), color='black', linestyle='--', linewidth=2, label=f'Mean: {nilai.mean():.1f}')
ax.axvline(np.median(nilai), color='orange', linestyle='--', linewidth=2, label=f'Median: {np.median(nilai):.1f}')

ax.set_title('Distribusi Nilai Ujian (n=500)', fontsize=14)
ax.set_xlabel('Nilai')
ax.set_ylabel('Frekuensi')
ax.legend()

plt.tight_layout()
plt.show()
```

## 4. Scatter Plot

```python
np.random.seed(42)
n = 200

jam_belajar = np.random.uniform(1, 10, n)
nilai_ujian = 50 + jam_belajar * 4 + np.random.normal(0, 5, n)
nilai_ujian = np.clip(nilai_ujian, 0, 100)
jurusan = np.random.choice(['IPA', 'IPS', 'Bahasa'], n)

fig, ax = plt.subplots(figsize=(9, 6))

color_map = {'IPA': '#2196F3', 'IPS': '#4CAF50', 'Bahasa': '#FF9800'}
for jur, color in color_map.items():
    mask = jurusan == jur
    ax.scatter(jam_belajar[mask], nilai_ujian[mask],
               c=color, label=jur, alpha=0.7, s=60, edgecolors='white', linewidths=0.5)

# Trend line
z = np.polyfit(jam_belajar, nilai_ujian, 1)
p = np.poly1d(z)
x_line = np.linspace(1, 10, 100)
ax.plot(x_line, p(x_line), 'k--', linewidth=1.5, alpha=0.8, label='Tren')

corr = np.corrcoef(jam_belajar, nilai_ujian)[0, 1]
ax.text(0.05, 0.95, f'Korelasi: {corr:.3f}',
        transform=ax.transAxes, fontsize=11,
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

ax.set_title('Hubungan Jam Belajar vs Nilai Ujian', fontsize=14)
ax.set_xlabel('Jam Belajar per Hari')
ax.set_ylabel('Nilai Ujian')
ax.legend()

plt.tight_layout()
plt.show()
```

## 5. Subplots — Multiple Plot dalam Satu Figure

```python
# 2x2 grid subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Dashboard Analisis Data', fontsize=16, fontweight='bold')

np.random.seed(42)
data = np.random.randn(200)

# Plot 1: Histogram
axes[0, 0].hist(data, bins=25, color='#2196F3', edgecolor='white')
axes[0, 0].set_title('Distribusi')

# Plot 2: Line
t = np.linspace(0, 4*np.pi, 200)
axes[0, 1].plot(t, np.sin(t), color='#4CAF50')
axes[0, 1].set_title('Gelombang Sin')

# Plot 3: Scatter
x = np.random.randn(100)
y = x + np.random.randn(100) * 0.5
axes[1, 0].scatter(x, y, alpha=0.6, color='#FF9800')
axes[1, 0].set_title('Scatter')

# Plot 4: Bar
kategori = ['A', 'B', 'C', 'D']
nilai_bar = [23, 45, 31, 18]
axes[1, 1].bar(kategori, nilai_bar, color=['#2196F3', '#4CAF50', '#FF9800', '#F44336'])
axes[1, 1].set_title('Bar Chart')

plt.tight_layout()
plt.show()
```

## 6. Pandas Built-in Plotting

```python
# Pandas punya .plot() langsung dari DataFrame
df = pd.DataFrame({
    'Q1': [120, 80, 95],
    'Q2': [135, 102, 115],
    'Q3': [148, 118, 130],
    'Q4': [165, 140, 155]
}, index=['Produk A', 'Produk B', 'Produk C'])

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Line plot
df.T.plot(ax=axes[0], marker='o', title='Penjualan per Kuartal')
axes[0].set_xlabel('Kuartal')

# Bar plot
df.plot(kind='bar', ax=axes[1], title='Perbandingan Produk', rot=0)

plt.tight_layout()
plt.show()
```

## 7. Simpan Plot

```python
# Buat plot
fig, ax = plt.subplots(figsize=(8, 5))
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x), color='#2196F3', linewidth=2)
ax.set_title('Plot yang Akan Disimpan')

# Simpan dalam berbagai format
fig.savefig('output_plot.png', dpi=150, bbox_inches='tight')
fig.savefig('output_plot.pdf', bbox_inches='tight')    # vector, kualitas terbaik
fig.savefig('output_plot.svg', bbox_inches='tight')    # scalable vector

plt.show()
print("Plot berhasil disimpan dalam 3 format!")

# Cleanup
import os
for f in ['output_plot.png', 'output_plot.pdf', 'output_plot.svg']:
    if os.path.exists(f):
        os.remove(f)
```
